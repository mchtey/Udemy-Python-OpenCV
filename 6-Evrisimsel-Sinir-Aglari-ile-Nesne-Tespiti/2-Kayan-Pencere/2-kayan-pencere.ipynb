## <font color=orange>**Kayan Pencere**</font> <font color=gray>**-**</font>

* Goruntu boyunca 'kayan' sabit genislik ve yukseklikte dikdortgen bir bolgedir.

* Bu pencerelerin her biri icin pencere bolgesi alinir ve pencerenin icerisinde bizi ilgilendiren bir nesnenin (bir yuze) olup olmadigini belirlemek icin bir goruntu siniflandirici uygulariz.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
def sliding_window(image, step, ws):
    
    for y in range(0, image.shape[0]-ws[1],step):
        for x in range(0, image.shape[1]-ws[0],step):
        
            yield(x,y, image[y:y+ws[1], x:x+ws[0]])
```

```python
img = cv2.imread('husky.jpg')
im = sliding_window(img, 50, (200,150))

for i, image in enumerate(im):
    if i == 18:
        print(image[0],image[1])
        plt.imshow(image[2])
```
