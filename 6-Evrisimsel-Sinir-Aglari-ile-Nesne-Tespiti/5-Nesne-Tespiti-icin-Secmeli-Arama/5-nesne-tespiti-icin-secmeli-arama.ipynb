## <font color=orange>**Nesne Tespiti icin Secmeli Arama**</font> <font color=gray>**-**</font>

* Secmeli arama, super piksel algoritmasi kullanarak bir gorseli asiri bolumlere ayirma yontemidir.

* Super piksel, ortak ozellikleri (piksel yogunlugu gibi) paylasan bir piksel grubu oalrak tanimlanabilir.

* Secmeli arama, bes temel benzerlik olcusune dayali olarak super pikselleri hiyerarsik bir sekilde birlestirir:

    * Renk benzerligi
    * Doku benzerligi
    * Boyut benzerligi
    * Sekil benzerligi
    * Yukaridaki benzerliklerin dogrusal kombinasyonu
    
* Secmeli arama, sinif etiketleri degil bolgeler olusturur. Olusturulan bu bolgeler sonrasinda siniflandiriciya girdi olarak verilecektir.

```python
import cv2
import random
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('pyramid.jpg')
image = cv2.resize(image,dsize=(600,600))

im = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(im),plt.axis('off');
```

```python
# ilklendir ss
ss = cv2.ximgproc.segmentation.createSelectiveSearchSegmentation()
ss.setBaseImage(image)
ss.switchToSelectiveSearchQuality()
```

```python
print('start')
rects = ss.process()
```

```python
output = im.copy()
```

```python
len(rects)
```

```python
# 9014 kutucuk olusmus. ilk 50'ye bakma:

for (x,y,w,h) in rects[:50]:
    color = [random.randint(0,255) for j in range(0,3)]
    cv2.rectangle(output, (x,y), (x+w, y+h), color, 2)
    
plt.imshow(output), plt.title('Output'),plt.axis('off');
```
