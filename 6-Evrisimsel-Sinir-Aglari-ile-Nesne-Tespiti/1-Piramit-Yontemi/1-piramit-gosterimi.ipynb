## <font color=orange>**Piramit Gosterimi**</font> <font color=gray>**-**</font>

* Goruntunun cok olcekli temsilidir.

* Goruntu piramidinin kullanilmasi, bÄ±r goruntunun farkli olceklerindeki goruntulerdeki nesneleri bulmamizi saglar.

* Piramidin altinda, orjinal boyutunda (genislik ve yukseklik acisindan) orjinal goruntu var ve sonraki her katmanda, goruntu yeniden boyutlandirilir (alt orneklenir). Sonra istege bagli olarak duzlestirilir. (Genellikle Gaussian bulaniklastirma yoluyla).

```python
import cv2
import matplotlib.pyplot as plt
```

```python
def image_pyramid(image, scale=1.5, minSize=(224,224)): 
    #minSize tuple olarak kullandik cunku goruntu iki boyutlu 
    yield image
    
    while True:
        w = int(image.shape[1]/scale)
        image = cv2.resize(image, dsize=(w,w))
        
        if (image.shape[0] < minSize[1]) or (image.shape[1] < minSize[0]):
            break  # kucultmeyi sonlandirma kosulu
            
        yield image
```

```python
img = cv2.imread('husky.jpg')
im = image_pyramid(img)
for i, image in enumerate(im):
    print(i)
```
