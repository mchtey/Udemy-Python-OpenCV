## <font color=orange>**Kenar Algilama**</font>

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread('london.jpg',0)
plt.imshow(img,cmap='gray')
plt.axis('off')
plt.title('London');
```

```python
#Kenar islemleri:
edges = cv2.Canny(image = img, threshold1=0,threshold2=255)
plt.figure(), plt.imshow(edges,cmap='gray'), plt.axis('off');
```

```python
med_val = np.median(img)
print(med_val)
```

```python
mean_val = np.mean(img)
print(mean_val)
```

```python
low = int(max(0,(1 - 0.33)*med_val))     #alt threshold
high = int(min(255,(1 + 0.33)*med_val))  #ust threshold

print(low)
print(high)
```

```python
#Kenar islemleri:
edges = cv2.Canny(image = img, threshold1=low,threshold2=high)
plt.figure(), plt.imshow(edges,cmap='gray'), plt.axis('off');
```

```python
'''
Goruntude su yuzeyi de kenarlık olarak 
belirginlesmis bu yuzden blur uygulama denenecek
'''

#tum goruntuye uygulama:
blurred_img = cv2.blur(img,ksize=(3,3))
plt.figure(), plt.imshow(blurred_img, cmap='gray'), plt.axis('off');
```

```python
med_val = np.median(blurred_img)
print(med_val)
```

```python
low = int(max(0,(1 - 0.33)*med_val))     #alt threshold
high = int(min(255,(1 + 0.33)*med_val))  #ust threshold

print(low)
print(high)
```

```python
edges = cv2.Canny(image = blurred_img, threshold1=low,threshold2=high)
plt.figure(), plt.imshow(edges,cmap='gray'), plt.axis('off');
```

```python
#sudaki yuzeyde kenarliklar azalsa da daha da azaltmak gerekir.
#Bunun icin ksize degeriyle oynanir.

blurred_img = cv2.blur(img,ksize=(5,5))
med_val = np.median(blurred_img)

low = int(max(0,(1 - 0.33)*med_val))
high = int(min(255,(1 + 0.33)*med_val))

edges = cv2.Canny(image = blurred_img, threshold1=low,threshold2=high)
plt.figure(), plt.imshow(edges,cmap='gray'), plt.axis('off');
```
