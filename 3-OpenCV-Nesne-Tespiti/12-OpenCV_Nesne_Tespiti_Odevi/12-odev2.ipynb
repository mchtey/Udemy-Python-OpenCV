## <font color=gold>**Odev**</font> <font color=gray>**(2)**</font>

```python
# opencv, numpy kutuphanelerini ice aktarilmasi

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Resmi siyah beyaz olarak ice aktarilmasi ve cizdirilmesi:

img = cv2.imread('odev2.jpg',0)
plt.figure(), plt.imshow(img, cmap='gray'), plt.axis('off'), plt.title('Odev-2');
```

```python
# Edge detection : Kenarlarin tespiti:

kenarlar = cv2.Canny(image = img, threshold1 = 160, threshold2 = 255)
plt.figure(), plt.imshow(kenarlar, cmap='gray'), plt.axis('off'), plt.title('Edge Detection');
```

```python
# Yuz tespiti icin gerekli haar cascade'i ice aktarilmasi:
faceDetect = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
# Yuz tespiti yapip sonuclari gorsellestirilmesi:
img = cv2.imread('odev2.jpg',0)
rects = faceDetect.detectMultiScale(img)

for (x,y,w,h) in rects:
    cv2.rectangle(img, (x,y), (x+w,y+h),(255,255,255),5)
plt.figure(), plt.imshow(img, cmap='gray'), plt.axis('off'), plt.title('Face Detection');
```

```python
# Initalize the HOG insan tespiti algoritmamizi cagirmasi ve SVM'yi set edilmesi

hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
```

```python
# Resime insan tespiti algoritmasinin uygulanmasi ve gorsellestirilmesi:

(rects,weights) = hog.detectMultiScale(img, padding = (8,8), scale = 1.03)

for (xA, yA, xB, yB) in rects:
    cv2.rectangle(img, (xA,yA), (xB,yB), (255,0,0),2)
    
plt.figure(), plt.imshow(img, cmap='gray'), plt.axis('off'), plt.title('People Detection');
```
