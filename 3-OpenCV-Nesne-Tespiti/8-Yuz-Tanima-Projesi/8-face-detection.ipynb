## <font color=gold>**Yuz Algilama**</font> <font color=gray>**(Face Detection)**</font>

```python
import cv2
import matplotlib.pyplot as plt
```

```python
einstein = cv2.imread('einstein.jpg',0)
plt.figure(), plt.imshow(einstein,cmap='gray'), plt.axis('off');
```

```python
# Siniflandirici (Daha onceden egitilmis siniflandirici)
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
# Resmin uzerinde yuz olup olmadigini siniflandiriyor.
face_rect = face_cascade.detectMultiScale(einstein)

for (x,y,w,h) in face_rect:
    cv2.rectangle(einstein, (x,y), (x+w, y+h), (255,255,255),8)
plt.figure(), plt.imshow(einstein, cmap='gray'), plt.axis('off');
```

```python
# Barcelona Ornegi:

barce = cv2.imread('barcelona.jpg',0)
plt.figure(), plt.imshow(barce,cmap='gray'), plt.axis('off');
```

```python
face_rect = face_cascade.detectMultiScale(barce)

for (x,y,w,h) in face_rect:
    cv2.rectangle(barce, (x,y), (x+w, y+h), (255,255,255),8)
plt.figure(), plt.imshow(barce, cmap='gray'), plt.axis('off');
```

```python
# Hatali tespitler var. Parametreler ile oynayarak giderilebilir.

face_rect = face_cascade.detectMultiScale(barce,minNeighbors=7)

for (x,y,w,h) in face_rect:
    cv2.rectangle(barce, (x,y), (x+w, y+h), (255,255,255),5)
plt.figure(), plt.imshow(barce, cmap='gray'), plt.axis('off');
```

```python
# Video

cap = cv2.VideoCapture(0)

while True:
    ret,frame = cap.read()
    if ret:
        face_rect = face_cascade.detectMultiScale(frame,minNeighbors=7)
        for (x,y,w,h) in face_rect:
            cv2.rectangle(frame, (x,y), (x+w, y+h), (255,255,255),5)
        cv2.imshow('Face Detect',frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'): break
    
cap.release()
cv2.destroyAllWindows()
```
